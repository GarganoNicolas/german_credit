```python
import pandas as pd
data = pd.read_csv('german_credit_data.csv', sep=',')
data['Job'] = data['Job'].replace({
    0: 'unskilled and non-resident', 
    1: 'unskilled and resident', 
    2: 'skilled', 
    3: 'highly skilled'
})
```

# Prepare data for model

```python
data1 = data.replace({
    'Job': {
        'unskilled and non-resident': 0, 
        'unskilled and resident': 1, 
        'skilled': 2, 
        'highly skilled': 3
    },
    'Checking_account': {
        'little': 1, 
        'moderate': 2, 
        'other': 0, 
        'rich': 3
    },
    'Saving_accounts': {
        'other': 0, 
        'little': 1, 
        'quite rich': 3, 
        'rich': 4, 
        'moderate': 2
    },
    'Sex': {
        'male': 0, 
        'female': 1
    },
    'Housing': {
        'own': 2, 
        'free': 1, 
        'rent': 0
    },
    'Risk': {
        'good': 0, 
        'bad': 1
    }
})
status(data1)
```

```python
data_ohe = pd.get_dummies(data1, dtype=int)
status(data_ohe)
```

```python
data_ohe.to_csv('data_ohe.csv', index = False)
```
